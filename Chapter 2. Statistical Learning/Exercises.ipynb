```python
sns.pairplot(data=college.iloc[:,1:11] )
```

```python
college['Elite'] = college.Top10perc.apply(lambda x: 'Yes' if x>50 else 'No')
```

```python
plt.figure(figsize=(12,9))
sns.boxplot(data=college, x='Elite', y='Outstate')
```

```python
plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
sns.histplot(data=college, x='Grad.Rate', bins=4)

plt.subplot(2,2,2)
sns.histplot(data=college, x='Grad.Rate', bins=[0,20,40,60,80,100,120])

plt.subplot(2,2,3)
sns.histplot(data=college, x='S.F.Ratio')

plt.subplot(2,2,4)
sns.histplot(data=college, x='S.F.Ratio', bins=10, binrange=(0,30))
```
